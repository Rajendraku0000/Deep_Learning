```python
import numpy as np

import pandas as pd

df=pd.read_csv("C:\\Users\\rajen\\OneDrive\\Desktop\\data\\HeavyRain.csv")

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
```

```python
df[["month","day","previous_rainfall","rainfall"]]=mn.fit_transform(df[["month","day","previous_rainfall","rainfall"]])

df.head()
```

```python
df=df.drop(columns=["Unnamed: 0","date"])
```

```python
x=df[["month","day","previous_rainfall"]].values
print(x.ndim)
x
```

```python
y=df["rainfall"].values
print(y.ndim)
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x
```

```python
y
```

```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(64,activation="relu",input_shape=(x_train.shape[1],)))
model.add(Dense(32,activation="relu"))
model.add(Dense(1,activation="linear"))
```

```python
model.compile(optimizer="adam",loss="mean_squared_error",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=10,verbose=1)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,accuracy_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred,normalize=False)
```
