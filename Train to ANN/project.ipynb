```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\rajen\\OneDrive\\Desktop\\data\\Churn_Modelling.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df["Geography"]=le.fit_transform(df["Geography"])
df["Gender"]=le.fit_transform(df["Gender"])
df["Surname"]=le.fit_transform(df["Surname"])
```

```python
df
```

```python
x=df.drop(columns=["Exited"])
y=df[["Exited"]]
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=tf.keras.Sequential()

model.add(tf.keras.layers.Dense( units=64,activation="relu",input_shape=(x_train.shape[1],)))

model.add(tf.keras.layers.Dense(units=128,activation="relu"))
model.add(tf.keras.layers.Dense(units=64,activation="relu"))
model.add(tf.keras.layers.Dense(units=1,activation="linear"))

model.compile(optimizer="adam",loss="mse",metrics=["mae"])
```

```python
x_train
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=10)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred_new=np.where(y_pred>=0.5,1,0)
y_pred_new
```

```python
confusion_matrix_new=confusion_matrix(y_test,y_pred_new)
```
